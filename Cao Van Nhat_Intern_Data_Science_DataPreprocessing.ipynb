```python
import os
import shutil
def move_file( src, destination, array):
    ##src: nguá»“n
    ##des: Ä‘Ã­ch
    ##array: máº£ng chá»©a item cáº§n di chuyá»ƒn vd: Tshirt
    #src='E:\\Data\\archive\\fashion-dataset\images'
    src_files = os.listdir(src)
    src_files= np.array(src_files)
    print(type(src_files))
    for item in array:
        for file_name in src_files:
            if item == file_name:
                full_file_name = os.path.join(src, file_name)
                shutil.copy(full_file_name, destination)
        print("Xong item")
```

```python
import cv2
global X
X=[]
global Y
Y=[]
```

```python
def applyData(listdir, X, Y):
    percent = 0
    for folder in os.listdir(listdir):
        curr_path = os.path.join(listdir, folder)
        for file in os.listdir(curr_path):
            curr_file = os.path.join(curr_path, file)
            image = cv2.imread(curr_file)
            X.append(image)
            Y.append(folder)
        percent = percent+1
        print(percent + "%")
```

```python
listdir = "E:\\Dataset\\fashion-dataset\\Divide_data"
applyData(listdir,X,Y)
```
